## Dataset Metadata Explanation (Roman Hindi)

Is dataset me wildfire events se related features diye gaye hain jo **fire ke first 5 hours ke behaviour** ko describe karte hain. In features ka use karke hume predict karna hai ki fire **evacuation zone ke 5 km ke andar kab tak pahunch sakti hai**.

### Identifier

- **event_id**  
  Har wildfire event ka unique identifier hai. Yeh sirf identification ke liye use hota hai, modeling ke liye nahi.

### Target Variables

- **time_to_hit_hours**  
  Yeh batata hai ki fire ko evacuation zone ke 5 km ke andar pahunchne me kitna time laga (hours me).

- **event**  
  Binary indicator hai.  
  - 1 → Fire 72 hours ke andar evacuation zone ke paas pahunch gayi  
  - 0 → Fire 72 hours ke andar waha nahi pahunchi

### Temporal Coverage Features

Yeh features batate hain ki **first 5 hours me fire monitoring ka data kitna available hai**.

- **num_perimeters_0_5h**  
  First 5 hours me kitni baar fire perimeter record hua.

- **dt_first_last_0_5h**  
  First aur last perimeter observation ke beech ka time difference.

- **low_temporal_resolution_0_5h**  
  Agar observation bahut kam hai ya interval 0.5 hour se kam hai to value 1 hoti hai.

### Growth Features

Yeh features fire ke **size aur growth pattern** ko represent karte hain.

- **area_first_ha**  
  Fire ka initial area hectares me.

- **area_growth_abs_0_5h**  
  First 5 hours me fire area me absolute growth.

- **area_growth_rel_0_5h**  
  Relative growth rate.

- **area_growth_rate_ha_per_h**  
  Fire kitni speed se grow ho rahi hai (hectares per hour).

- **log1p_area_first**  
  Initial area ka log transformation.

- **log1p_growth**  
  Growth ka log transformed version.

- **log_area_ratio_0_5h**  
  Area ratio ka log value.

- **relative_growth_0_5h**  
  Relative fire expansion.

- **radial_growth_m**  
  Fire radius me growth (meters me).

- **radial_growth_rate_m_per_h**  
  Radius growth ki speed.

### Centroid Kinematics Features

Yeh fire ke **movement direction aur speed** ko describe karte hain.

- **centroid_displacement_m**  
  Fire centroid kitna move hua.

- **centroid_speed_m_per_h**  
  Fire movement speed.

- **spread_bearing_deg**  
  Fire spread direction degrees me.

- **spread_bearing_sin / spread_bearing_cos**  
  Direction ko circular form me represent karne ke liye sine aur cosine encoding.

### Distance Features

Yeh features fire aur evacuation zone ke **distance relationship** ko describe karte hain.

- **dist_min_ci_0_5h**  
  Fire aur nearest evacuation zone ke centroid ke beech minimum distance.

- **dist_std_ci_0_5h**  
  Distance variation.

- **dist_change_ci_0_5h**  
  Distance change over time.

- **dist_slope_ci_0_5h**  
  Distance change ka trend.

- **closing_speed_m_per_h**  
  Fire evacuation zone ki taraf kitni speed se badh rahi hai.

- **closing_speed_abs_m_per_h**  
  Closing speed ka absolute value.

- **projected_advance_m**  
  Fire ka projected movement toward zone.

- **dist_accel_m_per_h2**  
  Distance change ki acceleration.

- **dist_fit_r2_0_5h**  
  Distance change model ka goodness of fit.

### Directionality Features

Yeh fire movement aur evacuation zone direction ke alignment ko batate hain.

- **alignment_cos**  
  Direction alignment cosine value.

- **alignment_abs**  
  Absolute alignment measure.

- **cross_track_component**  
  Sideways movement component.

- **along_track_speed**  
  Fire movement speed towards evacuation zone direction.

### Temporal Metadata

Yeh fire ignition ka **time related information** batate hain.

- **event_start_hour**  
  Fire kis hour me start hui (0–23).

- **event_start_dayofweek**  
  Week ka kaunsa din (0 = Monday, 6 = Sunday).

- **event_start_month**  
  Fire kis month me start hui (1–12).

Overall, yeh metadata wildfire ke **growth, movement, distance aur timing behaviour** ko capture karta hai, jisse early stage me fire risk ko estimate kiya ja sakta hai.
